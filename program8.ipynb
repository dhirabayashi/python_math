```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2*x + 3

x = np.arange(-3, 4, 0.05)
y = f(x)

plt.plot(x, y, "o")
plt.grid(color='0.8')
plt.show()
```

```python
def func_proc(x):
    return 2*x

x = np.arange(-3, 4)
y = func_proc(x)

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
def func_proc(x):
    return 2/x

x = np.arange(-6, 6)
y = func_proc(x)

plt.plot(x, y)
plt.grid(color='0.8')
plt.show()
```

```python
import sympy as sp
sp.init_printing()

# f(x) = x^2
x = sp.Symbol('x')
fx = x**2
display(fx)

# 
sp.plot(fx)
```

```python
sp.diff(fx)
```

```python
sp.diff(fx).subs(x, 3)
```

```python
# f(x) = x^2
x = sp.Symbol('x')
fx = x**2
display(fx)

# 
x1 = 3
y1 = fx.subs(x, x1)

# x1
a = sp.diff(fx).subs(x, x1)
display(sp.diff(fx))

# (x1, y1)a
y = a * (x - x1) + y1
display(y)

# 
sp.plot(fx, y)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random

# 
dx = [28, 26, 28, 27, 27, 20, 26, 22, 23, 19, 26, 23, 25, 21, 20, 18, 24, 19, 24, 25]          # 
dy = [111, 97, 102, 105, 108, 74, 116, 92, 112, 88, 116, 101, 93, 74, 87, 71, 94, 67, 105, 99] # 

# 
a = 0.0  # 
b = 0.0  # 

# y=0
min_res = 0.0
for i in range(20):
    y = a * dx[i] + b
    min_res += (dy[i]-y)**2
print(min_res)

dat=[]  # 
for i in range(500000):
    # 
    wa = (random.random() - 0.5) * 0.001
    wb = (random.random() - 0.5) * 0.001
    
    # 
    res = 0
    for j in range(20):
        y = (a + wa) * dx[j] + (b + wb)
        res += (dy[j] - y)**2
        
    # 
    if res < min_res:
        min_res = res    #  
        a = a + wa       # 
        b = b + wb       # 
    dat.append(min_res)  # 
print(a, b, min_res)

# 
plt.figure(figsize=(12,4))
plt.plot(dat)
plt.show()
```
