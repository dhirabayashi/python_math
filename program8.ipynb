```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return 2*x + 3

x = np.arange(-3, 4, 0.05)
y = f(x)

plt.plot(x, y, "o")
plt.grid(color='0.8')
plt.show()
```

```python
# f(x) = x^2
x = sp.Symbol('x')
fx = x**2
display(fx)

# 接点の座標
x1 = 3
y1 = fx.subs(x, x1)

# 微分係数（x1における接線の傾き）
a = sp.diff(fx).subs(x, x1)
display(sp.diff(fx))

# 接線の式（(x1, y1)を通って傾きがaの直線）
y = a * (x - x1) + y1
display(y)

# グラフを描画
sp.plot(fx, y)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import random

# データ
dx = [28, 26, 28, 27, 27, 20, 26, 22, 23, 19, 26, 23, 25, 21, 20, 18, 24, 19, 24, 25]          # 気温
dy = [111, 97, 102, 105, 108, 74, 116, 92, 112, 88, 116, 101, 93, 74, 87, 71, 94, 67, 105, 99] # 販売数

# 初期値
a = 0.0  # 直線の傾き
b = 0.0  # 切片

# y=0との「ずれ」の二乗和
min_res = 0.0
for i in range(20):
    y = a * dx[i] + b
    min_res += (dy[i]-y)**2
print(min_res)

dat=[]  # 「ずれ」の二乗和の最小値
for i in range(500000):
    # 傾きと切片の更新量を決定
    wa = (random.random() - 0.5) * 0.001
    wb = (random.random() - 0.5) * 0.001
    
    # 「ずれ」の二乗和
    res = 0
    for j in range(20):
        y = (a + wa) * dx[j] + (b + wb)
        res += (dy[j] - y)**2
        
    # 値を更新
    if res < min_res:
        min_res = res    # 仮の最小値 
        a = a + wa       # 傾き
        b = b + wb       # 切片
    dat.append(min_res)  # 最小値を追加
print(a, b, min_res)

# グラフを表示
plt.figure(figsize=(12,4))
plt.plot(dat)
plt.show()
```
