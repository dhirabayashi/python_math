```python
def func_6a(x):
    return 2 * x

def func_6b(x):
    return 2 * x + 3

x = []
y1 = []
y2 = []
for i in range(-3, 4):
    x.append(i)
    y1.append(func_6a(i))
    y2.append(func_6b(i))
    
plt.plot(x, y1)
plt.plot(x, y2)
#plt.axis('equal')
plt.grid(color='0.8')
plt.show()
```

```python
import numpy as np

def func_7a(x):
    return 2 * x
    
def func_7b(x):
    return 2 * x - 9

x1 = np.arange(-1.5, 2)
y1 = func_7a(x1)
plt.plot(x1, y1)

x2 = np.arange(3, 7)
y2 = func_7b(x2)
plt.plot(x2, y2)

plt.grid(color='0.8')
plt.show()
```

```python
def func_8a(x):
    return 2 * x

def func_8b(x):
    return 2 * (x - 4.5)

x1 = np.arange(-1.5, 2)
y1 = func_8a(x1)
plt.plot(x1, y1)

x2 = x1 + 4.5
y2 = func_8b(x2)
plt.plot(x2, y2)

plt.grid(color='0.8')
plt.show()
```
